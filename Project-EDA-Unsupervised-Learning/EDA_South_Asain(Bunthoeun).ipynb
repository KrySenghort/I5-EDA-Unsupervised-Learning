#### 4. Details each featurs

- **Country:** The name of the South Asian country being analyzed (e.g., India, Pakistan, Bangladesh).
- **Year:** The calendar year for which the data is recorded.
- **GDP (current US\$)**: The total market value of all goods and services produced within a country in a given year, expressed in current U.S. dollars without adjusting for inflation.
- **GDP growth (annual \%):** The annual percentage change in the GDP, indicating how much the economy has grown or shrunk compared to the previous year.
- **GDP per capita (current US\$):** GDP divided by the total population, showing the average economic output per person in current U.S. dollars.
- **Unemployment, total (% of total labor force) (modeled ILO estimate):** The percentage of the labor force that is unemployed, based on estimates from the International Labour Organization.
- **Inflation, consumer prices (annual %):** The annual percentage change in consumer prices, reflecting the rate at which the cost of goods and services increases or decreases.
- **Foreign direct investment, net inflows (% of GDP):** The net amount of foreign investments flowing into the country as a percentage of its GDP.
- **Trade (% of GDP):** The total value of imports and exports as a percentage of GDP, indicating the country's trade openness.
- **Gini index:** A measure of income inequality within a country, where 0 represents perfect equality and 100 represents maximum inequality.
- **Population, total :** The total number of people living in the country.
- **Population growth (annual %):** The annual percentage change in the total population.
- **Poverty headcount ratio at \$2.15 a day (2017 PPP) (% of population):** The percentage of the population living on less than \$2.15 a day (adjusted for purchasing power parity, PPP), a measure of extreme poverty.
- **Life expectancy at birth, total (years):** The average number of years a newborn is expected to live, assuming current mortality rates.
- **Mortality rate, infant (per 1,000 live births):** The number of infant deaths per 1,000 live births in a given year.
- **Literacy rate, adult total (% of people ages 15 and above):** The percentage of people aged 15 and above who can read and write.
- **School enrollment, primary (% gross):** The total primary school enrollment as a percentage of the population eligible for primary education.
- **Urban population (% of total population):** The percentage of the total population living in urban areas.
- **Access to electricity (% of population):** The percentage of the population with access to electricity.
- **People using at least basic drinking water services (% of population):** The percentage of the population with access to at least basic water services.
- **People using at least basic sanitation services (% of population):** The percentage of the population with access to basic sanitation facilities.
- **Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita):** The per capita CO2 emissions, excluding land use, land-use change, and forestry (LULUCF).
- **PM2.5 air pollution, mean annual exposure (micrograms per cubic meter):** The mean annual concentration of fine particulate matter (PM2.5) in the air, measured in micrograms per cubic meter.
- **Renewable energy consumption (% of total final energy consumption):** The percentage of total energy consumed that comes from renewable sources.
- **Forest area (% of land area):** The percentage of the country’s total land area covered by forests.
- **Control of Corruption: Percentile Rank:** A governance indicator measuring perceptions of corruption in the public sector, ranked as a percentile.
- **Political Stability and Absence of Violence/Terrorism: Estimate:** A governance indicator reflecting the likelihood of political instability and/or violence, such as terrorism.
- **Regulatory Quality: Estimate:** A governance indicator that measures the ability of the government to formulate and implement sound policies.
- **Rule of Law: Estimate:** A governance indicator measuring the extent to which agents have confidence in and abide by societal rules, including contract enforcement, property rights, and the judiciary.
- **Voice and Accountability: Estimate:** A governance indicator that reflects the degree to which citizens can participate in  selecting their government and the extent of freedom of expression, association, and media.
- **Individuals using the Internet (% of population):** The percentage of the population with access to the internet.
- **Research and development expenditure (% of GDP):** The percentage of GDP spent on research and development activities.
- **High-technology exports (% of manufactured exports):** The share of high-tech products in the country's total manufactured exports, indicating technological advancement.

#### 6. Handle missing values and Check missing values:
#### 6.1.  Replace placeholders ('..') with NaN and convert columns to appropriate types

```python
df_South_asain.replace('..', pd.NA, inplace=True)

# Convert all applicable columns to numeric:
for column in df_South_asain.columns:
    try:
        df_South_asain[column] = pd.to_numeric(df_South_asain[column], errors='ignore')
    except:
        pass

missing_values_summary = df_South_asain.isnull().sum().sort_values(ascending=False)
missing_values_summary[missing_values_summary > 0]
```

```python
# Drop columns with more than 50% missing data
threshold = 0.5 
columns_to_drop = missing_values_summary[missing_values_summary > len(df_South_asain) * threshold].index
df_cleaned = df_South_asain.drop(columns=columns_to_drop)

# Impute missing values 
numeric_columns = df_cleaned.select_dtypes(include=['number']).columns
df_cleaned[numeric_columns] = df_cleaned[numeric_columns].fillna(df_cleaned[numeric_columns].median())

# Confirm no missing values remain in numeric columns
missing_after_imputation = df_cleaned.isnull().sum().sort_values(ascending=False)
missing_after_imputation[missing_after_imputation > 0]
```
